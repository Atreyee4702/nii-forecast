```python
def adftest(df):
  is_stationary = True
  for col in df.columns.values:
    test_result = adfuller(df[col], autolag="AIC")
    if test_result[1] > 0.05:
      is_stationary = False
    print(col, test_result[1] < 0.05)

  return is_stationary
```
